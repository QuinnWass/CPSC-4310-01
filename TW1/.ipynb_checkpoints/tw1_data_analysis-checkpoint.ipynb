## Advertising data

Let's take a look at some data, ask some questions about that data, and then use linear regression to answer those questions!

A toy dataset for your exploratory data analysis:

./data/Advertising.csv

```python
import pandas as pd
# read CSV file stored in the current directory and save the results
data = pd.read_csv('./data/Advertising.csv', skipinitialspace=True, index_col=0)

# display the first 5 rows
data.head()
```

What are the features?
- **TV:** advertising dollars spent on TV for a single product in a given market (in thousands of dollars)
- **Radio:** advertising dollars spent on Radio
- **Newspaper:** advertising dollars spent on Newspaper

What is the response?
- **Sales:** sales of a single product in a given market (in thousands of items)

What else do we know?
- Because the response variable is continuous, this is a **regression** problem.
- There are 200 **observations** (represented by the rows), and each observation is a single market.

```python
print("The mean for TV advertising is:" ,data['TV'].mean())
print("The median for TV advertising is:" ,data['TV'].median())
print("The modes for TV advertising are:")
for mode in data['TV'].mode():
    print(mode)
print('\n')

print("The mean for Radio advertising is:" ,data['Radio'].mean())
print("The median for Radio advertising is:" ,data['Radio'].median())
print("The modes for Radio advertising are:")
for mode in data['Radio'].mode():
    print(mode)
print('\n')

print("The mean for Newspaper advertising is:" ,data['Newspaper'].mean())
print("The median for Newspaper advertising is:" ,data['Newspaper'].median())
print("The modes for Newspaper advertising are:")
for mode in data['Newspaper'].mode():
    print(mode)
print('\n')

print("The mean for Sales is:" ,data['Sales'].mean())
print("The median for Sales is:" ,data['Sales'].median())
print("The modes for Sales are:")
for mode in data['Sales'].mode():
    print(mode)
print('\n')
```

```python
print("Five number Summaries:")
print(data['TV'].describe()[3:])
print(data['Radio'].describe()[3:])
print(data['Newspaper'].describe()[3:])
print(data['Sales'].describe()[3:])
```

```python
print(data.boxplot())
```

The values in 3 reflect the values in 2. The Lower arm reflects the lower 25%, the box represents the middle 50% and the upper arm represents the upper 25% which is labeled as 75%. The numbers in 2 reflect the starting values while the boxplot shows the whole thing. Additionally, the boxplot shows outliers.
